## Recomendador simple

### El recomendador simple va a recomendar los elementos top a partir de una cierta métrica

```python
# Encontramos en los datos que hay una columan con puntuaciones y parece que es la opción obvia
```

Para hacer un recomendador simple hay que hacer la media de pesos de un criterio. Hay una fórmula y esta fórmula se va a traducir en una función

v = número de votos por peli

m = mínimo de votos que hacen falta (no dado)

R = nota media

c = voto medio 

```python
c = data['vote_average'].mean()
```

```python
c
```

```python
m = data['vote_count'].quantile(0.90)   #que esté por encima del quartil 90 en número de votos
m
```

```python
q_movies = data.copy().loc[data['vote_count'] >= m]    #Copy es para asegurarnos el nuevo df es independiente
q_movies.shape
```

```python
q_movies.head()
```

```python
def weighted_rating(x, m=m, c=c):     #peso de las películas
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * c)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating, axis=1)
```

```python
q_movies = q_movies.sort_values('score', ascending=False)   #según puntuación
```
